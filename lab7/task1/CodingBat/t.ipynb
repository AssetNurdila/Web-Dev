```python
#String-2
def double_char(str):
    return ''.join([c*2 for c in str])
```

```python
#List-1
def first_last6(nums):
    return nums[0] == 6 or nums[-1] == 6
```

```python
#List-1
def count_evens(nums):
    return sum(1 for x in nums if x % 2 == 0)
```

```python
#List-2
def sum13(nums):
    total = 0
    skip = False
    for x in nums:
        if x == 13:
            skip = True
        elif skip:
            skip = False
        else:
            total += x
    return total
```
