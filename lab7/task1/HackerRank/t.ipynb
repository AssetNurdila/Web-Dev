```python
def is_leap(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

year = int(input())
print(is_leap(year))
```

```python
x = int(input())
y = int(input())
z = int(input())
n = int(input())

print([[i, j, k] for i in range(x+1)
                     for j in range(y+1)
                     for k in range(z+1)
                     if i + j + k != n])
```

```python
students = []

for _ in range(int(input())):
    name = input()
    score = float(input())
    students.append([name, score])

scores = sorted(set(s[1] for s in students))
second = scores[1]

names = sorted([s[0] for s in students if s[1] == second])

for name in names:
    print(name)
```
